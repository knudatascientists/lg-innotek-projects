```python
import numpy as np
import pandas as pd
```

```python
DIR1='./l.csv'
DIR2='./m.csv'
DIR3='./n.csv'
DIR4='./o.csv'
DIR5='./p.csv'
DIR6='./q.csv'
```

```python
df1=pd.read_csv(DIR1, header=None)
df2=pd.read_csv(DIR2, header=None)
df3=pd.read_csv(DIR3, header=None)
df4=pd.read_csv(DIR4, header=None)
df5=pd.read_csv(DIR5, header=None)
df6=pd.read_csv(DIR6, header=None)
```

```python
df1.drop_duplicates(subset=0, keep='last', inplace=True)
df2.drop_duplicates(subset=0, keep='last', inplace=True)
df3.drop_duplicates(subset=0, keep='last', inplace=True)
df4.drop_duplicates(subset=0, keep='last', inplace=True)
df5.drop_duplicates(subset=0, keep='last', inplace=True)
df6.drop_duplicates(subset=0, keep='last', inplace=True)
```

```python
df1=df1.set_index(0)
df2=df2.set_index(0)
df3=df3.set_index(0)
df4=df4.set_index(0)
df5=df5.set_index(0)
df6=df6.set_index(0)
```

```python
# 1. Log 파일 합치기
dfall=pd.concat([df1,df2,df3,df4,df5,df6], join='inner', axis=1) # ignore_index= 'bool'
```

```python
# 2. Log들을 바코드 순(사전 순)으로 정렬
dfall=dfall.sort_values([0], ascending=True)
```

```python
dfall.iloc[1:,:] = dfall.iloc[1:,:].astype('float')
```

```python
row = dfall.shape[0]
col = dfall.shape[1]
```

```python
dfall[' '] = np.nan
dfall.iloc[2,290] = 'result'
dfall.iloc[:2,290] = ' '
```

```python
for i in range(row-3):
    res = []
    for j in range(col-1):
        a = dfall.iloc[i+3,j]
        high = dfall.iloc[1,j]
        low = dfall.iloc[2,j]
    
        if a > high or a < low:
            res.append(dfall.iloc[0,j])
        else:
            pass
    if res == []:
        dfall.iloc[i+3, 290] = "pass"
    else:
        dfall.iloc[i+3, 290] = ' '.join(res)
```
